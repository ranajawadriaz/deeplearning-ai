COSINE SIMILARITY

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentence_1="This is my love"
sentence_2="my love is this"
sentence_3="I like Salad"
```

```python
emb_1=embeddingModel.encode([sentence_1])
emb_2=embeddingModel.encode([sentence_2])
emb_3=embeddingModel.encode([sentence_3])
```

```python
vec_1=[emb_1[0].tolist()]
vec_2=[emb_2[0].tolist()]
vec_3=[emb_3[0].tolist()]
```

```python
print(cosine_similarity(vec_1,vec_2))
print(cosine_similarity(vec_1,vec_3))
print(cosine_similarity(vec_2,vec_3))
```

DIFFERENCE BETWEEN WORD BY WORD EMBEDDING AND SENTENCE EMBEDDING

```python
sentence_11="The kids play in the park."
sentence_22="The play was for kids in the park."
```

```python
list_11=["kids","play","park"]
list_22=["play","kids","park"]
```

```python
embeddingsList_1=[emb.tolist() for emb in embeddingModel.encode(list_11)]
embeddingsList_2=[emb.tolist() for emb in embeddingModel.encode(list_22)]
```

```python
mean1=np.stack(embeddingsList_1).mean(axis=0).tolist()
mean1[:4]
```

```python
mean2=np.stack(embeddingsList_2).mean(axis=0).tolist()
mean2[:4]
```
